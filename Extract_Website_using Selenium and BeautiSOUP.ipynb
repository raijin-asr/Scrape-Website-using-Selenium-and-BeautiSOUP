```python
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
import time
```

```python

try:
    url = "https://www.yhdfa.vn/about-us"
    driver.get(url)
    time.sleep(5)

    # Extract all text content
    page_text = driver.find_element(By.TAG_NAME, "body").text
    print("Page Text:\n", page_text)

    # Extract image descriptions (alt text)
    images = driver.find_elements(By.TAG_NAME, "img")
    for i, img in enumerate(images, start=1):
        alt_text = img.get_attribute("alt")
        src = img.get_attribute("src")
        print(f"Image {i}:")
        print(f"  Alt Text: {alt_text}")
        print(f"  Source: {src}")

finally:
    # Close the WebDriver
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Set up WebDriver options
options = Options()
options.add_argument('--headless')  # Headless mode
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('start-maximized')
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

# Initialize WebDriver
driver = webdriver.Chrome(options=options)

```

```python
# Try loading the page
try:
    url = "https://www.yhdfa.vn/product/E/A01/SAD01-22?productModel="  # Change the URL to test
    driver.get(url)

    # Wait for the page to load
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.TAG_NAME, "body"))
    )

    # Extract all text content
    page_text = driver.find_element(By.TAG_NAME, "body").text
    print("Page Text:\n", page_text)

    # Extract image descriptions (alt text)
    images = driver.find_elements(By.TAG_NAME, "img")
    for i, img in enumerate(images, start=1):
        alt_text = img.get_attribute("alt")
        src = img.get_attribute("src")
        print(f"Image {i}:")
        print(f"  Alt Text: {alt_text}")
        print(f"  Source: {src}")

finally:
    # Close the WebDriver
    driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

# Fetch the webpage
url = "https://www.yhdfa.vn/product/E/A01/SAD01-22?productModel="
response = requests.get(url)

# Parse the HTML
soup = BeautifulSoup(response.text, "html.parser")

# Extract text content
page_text = soup.get_text()
print("Page Text:\n", page_text)

# Extract image descriptions
images = soup.find_all("img")
for i, img in enumerate(images, start=1):
    alt_text = img.get("alt")
    src = img.get("src")
    print(f"Image {i}:")
    print(f"  Alt Text: {alt_text}")
    print(f"  Source: {src}")
```
